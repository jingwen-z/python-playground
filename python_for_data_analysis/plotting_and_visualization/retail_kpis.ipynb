```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Volume of transactions**

```python
plt.figure(figsize=(9, 6))

x = np.arange(4)
volume_k = [12, 34, 31, 20]

plt.title('Total volume of 4 kinds of fruits in 2020', size=13)
plt.bar(x, volume_k, width=0.4, color='#177cb0')
plt.tick_params(labelsize=13)
plt.xticks(np.arange(4), ('apple', 'banana', 'cherry', 'durian'))
plt.xlabel('Product', size=13)
plt.ylabel('Volume (k)', size=13)

plt.show()
```

This plot describes the volume(k) for each fruit. Among four fruits, bananas’ sales are the largest (34k units), however, it seems that consumers don’t like apple that much.

**Turnover of transactions**

```python
plt.figure(figsize=(9, 6))

year_n_1 = [1.5, 3, 10, 13, 22, 36, 30, 33, 24.5, 15, 6.5, 1.2]
year_n = [2, 7, 14, 17, 20, 27, 30, 38, 25, 18, 6, 1]

plt.title('Monthly total turnover in year N vs year N-1', size=13)
plt.plot(np.arange(12), year_n_1, marker='.', color='#177cb0',
         label='year N-1', linestyle='--', alpha=0.5)
plt.plot(np.arange(12), year_n, marker='.', color='#177cb0', label='year N')

plt.tick_params(labelsize=13)
plt.xticks(np.arange(12), np.arange(1,13))
plt.xlabel('Month', size=13)
plt.ylabel('Turnover (K euros) of ice-cream', size=13)
plt.ylim(bottom=0)
plt.legend(fontsize=13)

plt.show()
```

This plot compares two-year turnover together. The peak for both year is in the summer, for year N, the peak is in August; however, the peak is reached in June in the year N-1, which is caused by the heatwave in June.

**Transactions amount**

```python
plt.figure(figsize=(9, 6))

trx_nb = [0, 5, 15, 13, 10, 14, 20, 22, 13, 15, 12, 18, 23, 20, 16, 12, 3, 0]

plt.title('Hourly transaction amount in a day', size=13)
plt.plot(np.arange(7, 25), trx_nb, marker='.', color='#177cb0')

plt.tick_params(labelsize=13)
plt.xticks(np.arange(7, 25), np.arange(7, 25))
plt.xlabel('Hour', size=13)
plt.ylabel('Transaction amount (K)', size=13)
plt.ylim(bottom=0)

plt.show()
```

According to this plot, it's obvious that the peak in a day is around noon and afterworking like 19h, because customers work in the morning and afternoon, they have time to shopping during the noon and afterworking. Moreover, there are 2 small peak around 9h and 16h, that might because customer come to buy their breakfast and classes over around 16h.

**Average volume per transaction**

```python
plt.figure(figsize=(9, 6))

vol_per_trx = [0.2, 0.3, 0.4, 0.7, 1.3, 1.5, 1.6, 1.7, 1.3, 1, 0.5, 0.1]

plt.title('Monthly average volume per transaction of ice-cream in a year', size=13)
plt.bar(np.arange(12), vol_per_trx, width=0.4, color='#177cb0')

plt.tick_params(labelsize=13)
plt.xticks(np.arange(12), np.arange(1,13))
plt.xlabel('Month', size=13)
plt.ylabel('Average volume of ice-cream\nper transaction', size=13)
plt.ylim(bottom=0)

plt.show()
```

This plot show the average volume per transaction of ice-cream. Logically, the volume increases with temperature's increasing.

**Average purchase value**

```python
fig, axarr = plt.subplots(nrows=1, ncols=1, figsize=(9, 6))

vol_per_trx = [8.5, 7, 10, 11, 10, 12.3, 13.2, 11.6, 13, 14, 12, 11]
val_per_trx = [20, 25, 30, 25, 33, 27, 24, 26, 25, 28, 20, 22]

axarr.set_title('Monthly average volume & value per transaction', size=13)
vol_bar = axarr.bar(np.arange(12), vol_per_trx, width=0.4, color='#177cb0')

axarr_2 = axarr.twinx()
val_line = axarr_2.plot(np.arange(12), val_per_trx, marker='.', color='#FF9415')

axarr.tick_params(labelsize=12)
axarr.set_xticks(np.arange(12))
axarr.set_xticklabels(np.arange(1, 13))
axarr.set_xlabel('Month', size=13)
axarr.set_ylabel('Average volume per transaction\n(k units)', size=13)
axarr.set_ylim(bottom=0, top=20)

axarr_2.tick_params(labelsize=12)
axarr_2.set_ylim(bottom=0, top=35)
axarr_2.set_ylabel('Average value per transaction\n(k euros)', size=13, rotation=270, labelpad=33)
axarr_2.legend((vol_bar[0], val_line[0]), ('Volume', 'Turnover'))

plt.show()
```

This plot describes the average volume and average value per sale during 12 months. We find that when the average volume per sale increases, the average value does not always increase. For instance, the average volume of April is more than March, but the average value is less, that means customers buy more units during their purchase, but the price per unit is decreasing.

**Purchase frequency**

```python
fig, axarr = plt.subplots(nrows=1, ncols=1, figsize=(9, 6))

freq = [10, 9, 6, 5, 5, 6, 7, 7, 9, 8, 6, 6]
val_per_trx = [20, 25, 50, 70, 78, 73, 65, 55, 60, 50, 53, 63]

axarr.set_title('Monthly purchase frequency per client\n& monthly average value per transaction in a year', size=13)
vol_bar = axarr.bar(np.arange(12), freq, width=0.4, color='#177cb0')

axarr_2 = axarr.twinx()
val_line = axarr_2.plot(np.arange(12), val_per_trx, marker='.', color='#FF9415')

axarr.tick_params(labelsize=12)
axarr.set_xticks(np.arange(12))
axarr.set_xticklabels(np.arange(1, 13))
axarr.set_xlabel('Month', size=13)
axarr.set_ylabel('Purchase frequency', size=13)
axarr.set_ylim(bottom=0, top=16)

axarr_2.tick_params(labelsize=12)
axarr_2.set_ylim(bottom=0, top=80)
axarr_2.set_ylabel('Average value per transaction\n(k euros)', size=13, rotation=270, labelpad=33)
axarr_2.legend((vol_bar[0], val_line[0]), ('Frequency', 'Turnover'))

plt.show()
```

Imagine that this plot shows the performance of some supermarkets of France in 2020. The purchase frequency reduces because of the lockdown for COVID-19, however, the average value per sale doubles. Thus, although the frequency reduces, the total turnover doesn't become less by analysing the average value per sale.

**Gross Margin Return On Investment (GMROI)**

```python
plt.figure(figsize=(9, 6))

cleaning_prod_gmroi = [0.6, 0.8, 1.2, 1.3, 1.5, 1.4, 1.2, 1.3, 1.2, 1.2, 1.1, 1]
staple_prod_gmroi = [0.8, 1, 1.3, 1.7, 1.4, 1.4, 1.1, 1, 1.4, 1.2, 1, 0.9]
fruits_gmroi = [0.5, 0.6, 1, 1.1, 0.9, 1, 1.1, 0.9, 1.2, 1, 1.1, 0.8]

plt.title('Monthly GMROI in a year', size=13)
plt.plot(np.arange(12), cleaning_prod_gmroi, marker='.', color='#177cb0', label='Cleaning products')
plt.plot(np.arange(12), staple_prod_gmroi, marker='.', color='#FF9415', label='Staple products')
plt.plot(np.arange(12), fruits_gmroi, marker='.', color='#10C062', label='Fruits')

plt.tick_params(labelsize=13)
plt.xticks(np.arange(12), np.arange(1,13))
plt.xlabel('Month', size=13)
plt.ylabel('Gross Margin Return On Investment', size=13)
plt.ylim(bottom=0)

plt.legend(fontsize=12, loc=3)
plt.show()
```

This plot describes the GMROI of 3 groups of products (cleaning products, staple products and fruits) of 2020. All of them increase since March because of the COVID-19 lockdown, customers were willing to buy more cleaning products to keep clean, to buy more staple products and fruits since they need to cook at home and to keep the dietary balance.

**Turnover per square foot**

```python
plt.figure(figsize=(9, 6))

x = np.arange(4)
turnover_k = [14, 26, 17, 15]

plt.title('Daily turnover per square foot of 4 kinds of products', size=13)
plt.bar(x, turnover_k, width=0.4, color='#177cb0')
plt.tick_params(labelsize=13)
plt.xticks(np.arange(4), ('Fruits', 'Wine', 'Butchery', 'Seafood'))
plt.xlabel('Product', size=13)
plt.ylabel('Turnover per square foot (k euros)', size=13)

plt.show()
```

This plot shows the turnover per square foot of 4 kinds of products during one day, the value for the wine is higher than others since they are put in the multi-layer shelves, which is a good example of using spaces. Fruits' turnover per square foot is the lowest among the four, because its unit price is lower, although its space utilization is better than the shelves of butchery and seafood. For the other two kinds of products, their space utilization is not enough, but their unit price is high, so their turnovers per square are not that low.

**Year over year growth**

```python
fig, axarr = plt.subplots(nrows=1, ncols=1, figsize=(9, 6))

year_n = [20, 18, 17, 19, 20, 21, 20, 22, 20, 19, 18, 21]
yoy = [0.3, 0.2, 0.18, 0.21, 0.19, 0.25, 0.17, 0.20, 0.16, 0.2, 0.18, 0.27]

axarr.set_title('Monthly total turnover & Year over year growth in year N', size=13)
yoy_bar = axarr.bar(np.arange(12), yoy, width=0.4, color='#177cb0')

axarr.tick_params(labelsize=12)
axarr.set_xticks(np.arange(12))
axarr.set_xticklabels(np.arange(1, 13))
axarr.set_xlabel('Month', size=13)
axarr.set_ylabel('Year over year growth', size=13)
axarr.set_ylim(bottom=0, top=0.4)

axarr_2 = axarr.twinx()
turnover_line = axarr_2.plot(np.arange(12), year_n, marker='.', color='#FF9415', label='year N')
axarr_2.tick_params(labelsize=12)
axarr_2.set_ylabel('Turnover (k euros)', size=13, rotation=270, labelpad=17)
axarr_2.set_ylim(bottom=0, top=25)

axarr_2.legend((yoy_bar[0], turnover_line[0]), ('Year over year growth', 'Turnover'), ncol=2, fontsize=12)

plt.show()
```

This plot shows the turnover and the year over year growth for each month, which helps us to identify if the increasing of turnover is also significant vs. the last year.

**Inventory turnover**

```python
fig, axarr = plt.subplots(nrows=1, ncols=1, figsize=(9, 6))

turnover = [2.6, 2.5, 3.0, 3.3, 3.5, 3.3, 3.1, 3.2, 3, 3.1, 2.9, 3]

axarr.set_title('Inventory turnover in year 2020', size=13)
axarr.bar(np.arange(12), turnover, width=0.4, color='#177cb0')

axarr.axhline(y=2.7, ls='--', color='#FF9415')
axarr.text(0.5, 2.8, 'Target', size=12, color='#FF9415')
axarr.tick_params(labelsize=12)
axarr.set_xticks(np.arange(12))
axarr.set_xticklabels(np.arange(1, 13))
axarr.set_xlabel('Month', size=13)
axarr.set_ylabel('Inventory turnover', size=13)
axarr.set_ylim(bottom=0, top=4)

plt.show()
```

The target inventory turnover of 2020 is 2.7, which means the net sales is twice more than the average inventory at selling price. Because of the lockdown, the sales had a high increasing. which brings a high inventory turnover.

**Sell-through rate**

```python
df = pd.DataFrame({'product': ['fruit', 'cheese', 'wine', 'vegetable', 'poultry', 'sushi', 'butchery'],
                   'Sell-through rate': [0.72, 0.7, 0.6, 0.7, 0.8, 0.95, 0.9]},
                  columns=['product', 'Sell-through rate'])
df.sort_values('Sell-through rate', inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
plt.figure(figsize=(9, 6))

plt.barh(np.arange(len(df['product'])), df['Sell-through rate'], align='center', height=0.6)
plt.yticks(np.arange(len(df['product'])), df['product'])
plt.tick_params(labelsize=13)
plt.xlabel('Sell-through rate', fontdict={'fontsize': 13})
plt.ylabel('Product', fontdict={'fontsize': 13})
plt.xlim(0, 1)

plt.show()
```

Among these 7 kinds of products, sushi's sell-through rate is the highest since they can nearly all be sold the same day; on the contrary, the sell-through rate of the wine is lower than others since it's not the daily aliment for everyone.
